### Split the dataset

```python
fv = data.iloc[:,-1]
fv
```

```python
cv = data.iloc[:,0]
cv
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(fv,cv,test_size=0.2,random_state=3,stratify=cv)
```

### Data Preprocessing

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
import re
```

```python
def lowers(x):
    return x.str.lower()
```

```python
def html(x):
    return x.apply(lambda x:re.sub("<.+?>"," ",x))
```

```python
def url(x):
    return x.apply(lambda x:re.sub("http[s]?://.+? +"," ", x))
```

```python
def unw(x):
    return x.apply(lambda x:re.sub("[]:()*\-.,@#$%^&0-9]"," ", x))
```

```python
txt_preprocess_pipeline = Pipeline([("lower",FunctionTransformer(lowers)),
         ("html",FunctionTransformer(html)),
         ("url",FunctionTransformer(url)),
         ("unw",FunctionTransformer(unw))])
```

```python
pp_x_train=txt_preprocess_pipeline.fit_transform(x_train)
pp_x_train
```

### Feature Engineering

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_on_pptext=Pipeline([("Pre-process",txt_preprocess_pipeline),("bagofwords",CountVectorizer())])
```

```python
bow_on_pptext
```

```python
bow_on_pptext.fit_transform(x_train)
```

### Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
b=BernoulliNB(alpha=1.0)
```

```python
final_model=b.fit(bow_on_pptext.fit_transform(x_train),y_train)
```

### Accuracy of the model


```python
from sklearn.metrics import accuracy_score
```

```python
predict=final_model.predict(bow_on_pptext.transform(x_test))
```

```python
accuracy_score(y_test,predict)
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
cm = confusion_matrix(y_test, predict)
cm
```

```python
ConfusionMatrixDisplay(cm)
```

```python
i=ConfusionMatrixDisplay(cm)
i.plot()
plt.show()
```

### Classification Report

```python
print(classification_report(y_test, predict))
```
